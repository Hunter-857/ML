```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

%matplotlib inline
```

```python
person = mpimg.imread('image/boy.png')
background = mpimg.imread('image/back.png')

plt.figure(figsize = (12,12))
plt.subplot(1,2,1)
plt.imshow(person)
plt.title("Person")
plt.subplot(1,2,2)
plt.imshow(background)
plt.title("back group")
```

```python
## 需要吧 矩阵搞成一样的
combo1 =  .5 * person + .5 * background
combo2 =  .75 * person + .25 * background
```

```python
## 2个图片尺寸不一样画不出来
plt.figure(figsize = (12,12))
plt.subplot(1,2,1)
plt.imshow(combo1)
plt.title("Person")
plt.subplot(1,2,2)
plt.imshow(combo2)
plt.title("back group")
```

```python
A = np.array([(.5,.5),(.25,.75)])
b = np.array((combo1,combo2),axis=2)
```

```python
out =np.linalg.solve(A,b)
out_background = out[:,:,0,:]
out_person = out[:,:,1,:]
```

```python
plt.imshow(out_background)
```

```python
plt.imshow(out_person)
```
